```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# Read the Excel files
df_3 = pd.read_excel("/content/metadonne-def-3.xlsx")
df_insecticide = pd.read_excel("/content/metadonnee-insecticides3-3.xlsx")
df_herbecide = pd.read_excel("/content/metaherbicide4.xlsx")

# Display the first few rows of each DataFrame
print("df_3:")
display(df_3.head())  # Use display() in Colab for better output

print("\ndf_insecticide:")
display(df_insecticide.head())

print("\ndf_herbecide:")
display(df_herbecide.head())
```

```python
# Print the number of rows in each DataFrame
print(f"Number of rows in df_3: {df_3.shape[0]}")
print(f"Number of rows in df_insecticide: {df_insecticide.shape[0]}")
print(f"Number of rows in df_herbecide: {df_herbecide.shape[0]}")
```

```python
print("Missing values in df_3:")
print(df_3.isnull().sum())

print("\nMissing values in df_insecticide:")
print(df_insecticide.isnull().sum())

print("\nMissing values in df_herbecide:")
print(df_herbecide.isnull().sum())
```

Merging the data


```python
df_3["type"] = "fongecide"
df_insecticide["type"] = "insecticide"
df_herbecide["type"] = "herbecide"

# Merge the datasets with the new column
df_all = pd.concat([df_3, df_insecticide, df_herbecide], ignore_index=True)

# Display the first few rows
print(df_all.head())
```

```python
# List of columns to drop
columns_to_drop = ["images", "source_ar", "unite_ar", "utilisation_ar", "frequence_ar", "composante_ar", "unite","frequence","utilisation"]

# Drop columns, ignoring errors if a column is missing
df_all = df_all.drop(columns=columns_to_drop, errors='ignore')

# Display the updated dataframe
print(df_all.head())
```

```python
print("number of rows")
print(df_all.shape[0])
print("number of columns")
df_all.isnull().sum()
```

```python
# Get unique values in the 'source' column
unique_sources = df_all["source"].unique()

# Display the unique sources
print(f"Unique sources in 'source': {unique_sources}")

# Alternatively, if you want the count of each source:
source_counts = df_all["source"].value_counts()

# Display the source counts
print(f"Source counts in 'source':\n{source_counts}")
```

```python
# Nettoyage des espaces avant et aprÃ¨s dans la colonne 'source'
df_all['source'] = df_all['source'].str.strip()
```

```python
# Compter le nombre d'occurrences de chaque source
source_counts = df_all['source'].value_counts()

# Afficher les rÃ©sultats
print(f"Source counts in 'source':\n{source_counts}")
```

```python
# Distribution des types
plt.figure(figsize=(10, 6))
sns.countplot(data=df_all, x='type', palette='viridis')
plt.title('Distribution of sources by type')
plt.xlabel('Type')
plt.ylabel('Product number')
plt.tight_layout()
plt.show()
```

```python
# Graphique en barres empilÃ©es par type et source
plt.figure(figsize=(12, 6))
df_all.groupby(['type', 'source']).size().unstack().plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Paired')
plt.title('Distribution of sources by type')
plt.xlabel('Type')
plt.ylabel('Number of products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Graphique en secteurs pour les sources
plt.figure(figsize=(8, 8))
source_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colormap='Set3')
plt.title('Distribution of sources')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
