```python
import pandas as pd

# Charger les fichiers Excel
df1 = pd.read_excel("/content/prix-herbicide3-1.xlsx")
df2 = pd.read_excel("/content/prixfongicide2-3-2.xlsx")
df3 = pd.read_excel("/content/prixinsecticides3-2.xlsx")

# Ajouter une colonne "Type" à chaque DataFrame
df1['Type'] = 'Herbicide'
df2['Type'] = 'Fongicide'
df3['Type'] = 'Insecticide'

# Concatenner les DataFrames (ajouter les lignes de df2 et df3 à df1)
df_combined = pd.concat([df1, df2, df3], ignore_index=True)

# Sauvegarder le DataFrame combiné dans un nouveau fichier Excel
df_combined.to_excel("/content/combined_prix_products.xlsx", index=False)

# Afficher les premières lignes pour vérifier
df_combined.head()
```

```python
total_lignes = len(df_combined)
print("Nombre total de lignes  :", total_lignes)
```

```python
# Appliquer la logique pour remplir la colonne 'prix'
df_combined['prix'] = df_combined['prix'].fillna(df_combined['prixmax']).fillna(df_combined['prixmin'])

# Supprimer les colonnes 'date', 'prixmin' et 'prixmax'
df_combined = df_combined.drop(columns=['date', 'datemin', 'prixmin', 'prixmax'])
df_combined.head()
```
