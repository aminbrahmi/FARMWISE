### Loading the Data
First, we load the dataset into a pandas DataFrame for analysis

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the data with the correct encoding
try:
    data = pd.read_csv('data diseases_description_recommanded_product.csv', sep=';', encoding='utf-8')
except UnicodeDecodeError:
    # If utf-8 fails, try latin1 or ISO-8859-1
    data = pd.read_csv('data diseases_description_recommanded_product.csv', sep=';', encoding='latin1')

# Display the first few rows to understand the data structure
data.head()
```

### Data Exploration
Let’s explore the dataset to understand its structure and content.

General information about the data

```python
data.info()

# Descriptive statistics
data.describe(include='all')
```

### Data Visualization

#### 1. Number of Recommended Products by Disease
This visualization shows how many products are recommended for each disease.

Visualization 1: Number of recommended products by disease

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Diseases', data=data, palette='viridis')
plt.title('Number of Recommended Products by Disease')
plt.xlabel('Number of Products')
plt.ylabel('Diseases')
plt.show()
```

#### 2. Distribution of Recommended Products
This visualization shows the frequency of use of different recommended products.

Visualization 2: Distribution of recommended products
Split recommended products (they are separated by commas)

```python
product_counts = data['Recommended Product'].str.split(',').explode().str.strip().value_counts()

plt.figure(figsize=(10, 6))
product_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Recommended Products')
plt.xlabel('Products')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Interpretation of Results

#### Key Insights
1. **Number of Recommended Products by Disease**:
   - Some diseases have multiple recommended products, while others may have only one. 
     This helps identify diseases that require more complex treatment strategies.

2. **Distribution of Recommended Products**:
   - Certain products, such as **Chlorothalonil** or **Copper-based products**, appear frequently in the recommendations. 
     These are likely the most commonly used treatments across various diseases.
