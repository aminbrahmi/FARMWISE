## Objective
This analysis aims to explore and visualize the price distribution of agricultural components based on their category (herbicides, fertilizers, etc.). We will use graphical representations to better understand price distribution and identify the most expensive components.

Now, let's load the data and start the analysis! ðŸš€

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le fichier CSV
file_path = "agri_data_prix_modified_data.csv"  # Adapte le chemin selon ton environnement
df = pd.read_csv(file_path)

```

# Price Distribution

```python
# DÃ©finir le style de Seaborn
sns.set(style="whitegrid")

# 1. Distribution des prix
plt.figure(figsize=(12, 6))
sns.histplot(df['prix'], bins=20, kde=True, color='blue')
plt.title("Distribution des prix des composants agricoles")
plt.xlabel("Prix")
plt.ylabel("FrÃ©quence")
plt.show()

```

# Price Distribution by Type

```python
# 2. RÃ©partition des prix par type
plt.figure(figsize=(12, 6))
sns.boxplot(x="Type", y="prix", data=df, palette="Set2")
plt.title("RÃ©partition des prix par type de composante")
plt.xticks(rotation=45)
plt.show()

```

# Top 10 Most Expensive Components

```python
# 3. Top 10 des composantes les plus chÃ¨res
top_10 = df.groupby("Composante")["prix"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10.values, y=top_10.index, palette="coolwarm")
plt.title("Top 10 des composantes les plus chÃ¨res (prix moyen)")
plt.xlabel("Prix moyen")
plt.ylabel("Composante")
plt.show()
```

# Top 10 Less Expensive Components

```python
# 3. Top 10 des composantes les plus chÃ¨res
top_10 = df.groupby("Composante")["prix"].mean().sort_values(ascending=True).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10.values, y=top_10.index, palette="coolwarm")
plt.title("Top 10 des composantes les plus chÃ¨res (prix moyen)")
plt.xlabel("Prix moyen")
plt.ylabel("Composante")
plt.show()
```
